# Leetcode 146 LRUCache

题目链接：https://leetcode.com/problems/lru-cache/

解题思路：
题目本身不难，get接口要求我们实现根据对应的key取出对应的value，这是唯一能够使用到key value的操作。

put接口要求我们向其中存入对应的key, value,当超出容量之后，将最近使用频率最少的key value移出

但是要求我们实现的接口的时间复杂度是常数，所以对数据结构有一定的要求。我能想到的是使用map，但是python中没有这个数据结构，可以用字典来代替，字典中对应的插入操作dict[key] = value, 删除操作del dict[key]的时间复杂度都是常数，符合题目要求。key value的使用频率对应于其在字典中的插入顺序，即序号靠前的key代表其使用频率少，序号靠后的key代表其使用频率高。

PS.在leetcode支持的python3版本中，不必使用OrderedDict(),dict本身就可以记住插入的顺序

在get()接口中，我们需要判断当前字典中是否有key，如果有，则将value取出，并将其删除后重新加入字典中，这属于更新操作，因为它被用到了，所以应该放在最后

在put接口中，我们首先判断其是否已经在字典中，若在，直接将其更新到字典末尾，不在，判断容量是否已满，不满，直接插入，满了，则将使用频率最少的元素删除（即key中最靠前的元素）

```python
class LRUCache:
    def __init__(self, capacity: int):
        self.cap = capacity
        self.cache = dict()

    def get(self, key: int) -> int:
        if key in self.cache:
            value = self.cache.pop(key)
            self.cache[key] = value
            return value
        return -1

    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            del self.cache[key]
        elif len(self.cache) == self.cap:
            del self.cache[list(self.cache.keys())[0]]

        self.cache[key] = value
```
