# Leetcode 124 Binary Tree Maximum Path Sum

题目链接：https://leetcode.com/problems/binary-tree-maximum-path-sum/


题目描述：

本题要求我们寻找一条最长路径，路径的根节点可以是二叉树上的任意节点

解题思路：

其实我们只需要考虑有可能是最长路径的情况，1.当前节点加上其左边最长边以及右边最长边 2.以左孩子节点为根节点的最长路径 3.以右孩子为根节点的最长路径。能够明确这三种情况就是解决问题的关键

代码如下：

```python
class Solution:
    def maxPathSum(self, root: TreeNode) -> int:
        self.max = root.val
        def maxSide(root):
            if root is None:
                return 0
            left_path = max(0, maxSide(root.left))
            right_path = max(0, maxSide(root.right))
            
            #update the max node while calculating the longest left/right path
            self.max = max(self.max, root.val + left_path + right_path)
            return max(left_path, right_path) + root.val

        maxSide(root)
        return self.max
```
