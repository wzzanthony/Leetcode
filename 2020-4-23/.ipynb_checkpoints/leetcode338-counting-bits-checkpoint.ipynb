# leetcode 338-counting-bits



本题要求我们计算0到num这个数之间所有数，他们以二进制形式表示，其中含有1的个数，以数组的形式存储并作为最后的返回结果，并且要求时间复杂度为$$O(n)$$,空间复杂度也为$$O(n)$$

解题思路：
既然要求时间复杂度为$$O(n)$$,这就意味着我们必然要使用到之前的结果，我们先写几个数，找找规律

|十进制形式|二进制形式|含有1的个数|
|----------|----------|-----------|
|0|0000|0|
|1|0001|1|
|2|0010|1|
|3|0011|2|
|4|0100|1|
|5|0101|2|
|6|0110|2|
|7|0111|3|
|8|1000|1|
|9|1001|2|
|10|1010|2|

显然，我们可以发现，所有的偶数，其二进制形式的最后一位是0，所以对应的+1之后的奇数所含有1的个数只要在其含有1的个数上加一即可
但是偶数的规律就不太容易发现了。其实我觉得我们可以借鉴一下从奇数上发现的规律，再发现一些更深入的规律，我们假设奇数为m，偶数为m-1，m含有1的个数之所以比m-1含有1的个数多一个，就是因为其最后一位为1，而m-1最后一位为0。这也就意味着，抛开最后一位来看，他们二者是一样的。抛开最后一位，可以通过右移操作来实现，并且这样我们就能用到之前的结果（之前的结果存在ret列表中），根据此数的奇偶性选择+1或者不加即可

代码如下：

```python
class Solution:
    def countBits(self, num: int) -> List[int]:
        ret = [0]
        for i in range(1, num+1):
            ret.append(ret[i >> 1] + i%2)
            
        return ret
```
