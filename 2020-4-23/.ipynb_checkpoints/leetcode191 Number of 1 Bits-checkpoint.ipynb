# Leetcode 191 Number of 1 Bits

题目链接：https://leetcode.com/problems/number-of-1-bits/



解题思路：此题要求比较简单，只需要进行移位操作（32次 int 32位），判断移出的是1还是0（奇偶性）

代码如下：

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        ret = 0
        for i in range(32):
            ret += n % 2
            n = n >> 1
        return ret
```
