# Leetcode 221 Maximal Square 

题目链接：https://leetcode.com/problems/maximal-square/



题目要求：
从给定的字符数组中，找出包含全1的最大方阵，并返回其面积

解题思路：
首先明确我们将使用动态规划的算法，更新的原则很简单，首先判断当前位置([i,j])是否为1，不是，直接跳过，无需更新。是1的话，则将其更新为
[i-1,j],[i,j-1],[i-1,j-1]三个对应位置中最小元素值加一（可以简单地理解为，当前位置所能形成的最大方阵，是由其周边所能形成的最小方阵决定的）。在更新二维数组的过程中，同时可以更新最大值（即方阵的边长，最终结果的根）。需要注意的一个细节是，初始时元素全部都为字符串，所以应当进行一步转换，将其转换为整数类型

例如：


代码如下：

```python
class Solution:
    def maximalSquare(self, matrix: List[List[str]]) -> int:
        if not matrix:
            return 0
        ret = 0
        for i in range(len(matrix)):
            matrix[i][0] = int(matrix[i][0])
            ret = max(ret, matrix[i][0])
        for j in range(len(matrix[0])):
            matrix[0][j] = int(matrix[0][j])
            ret = max(ret, matrix[0][j])
        
        for i in range(1, len(matrix)):
            for j in range(1, len(matrix[0])):
                matrix[i][j] = int(matrix[i][j])
                if matrix[i][j] == 1:
                    matrix[i][j] = min(matrix[i-1][j], matrix[i][j-1], matrix[i-1][j-1]) + 1
                    ret = max(matrix[i][j], ret)
        return ret**2
```
