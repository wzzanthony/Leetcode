# Leetcode 1143 Longest Common Subsequence

题目链接：https://leetcode.com/problems/longest-common-subsequence/

题目要求：

首先，我们要明确子串的定义。子串的中的元素是字符串中的元素，可以不连续，但是顺序要保持一致。

解题思路：

这一道题是动态规划中的经典问题，如果采用暴力解的方式，时间复杂度是$O(n * 2^n)$,简单地说，一个字符串有$2^n$个子串，之后，判断每个子串中的元素是否全部包含在另一个长度为n的字符串中即可。时间复杂度相当高，这种方法肯定不可取。


需要利用动态规划的方法来解决此问题。我们知道动态规划需要满足的条件之一就是最优子结构。我们先对此进行一个简单的证明，如果text1和text2的第一个元素相等，那么此问题的解就变成了求解text[1:]和text2[1:]的最长公共子序列（最终结果要加一）。这也就意味着，text1[0]必然是最长公共子序列中的一个元素。我们先假设text1[0]不是最长公共子序列中的元素，那我们就可以将两个字符串的第一个元素(因为相等)剔除，text1和text2最长公共子序列的解就是text1[1:]和text2[1:]最长公共子序列的解。假设这个解为s,又已知text1和text2的首元素相等,text1和text2的最长公共子序列的解应该是s+1而不是s。出现矛盾，所以如果首元素相等，那么其必定包含在最终解中。如果首元素不相等，那么问题就变成了寻找text1[1:]和text2的最长公共子序列和text1和text2[2:]的最长公共子序列中的最优解。

代码如下：

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        if len(text1) == 0 or len(text2) == 0:
            return 0
        if text1[0] == text2[0]:
            return 1 + self.longestCommonSubsequence(text1[1:], text2[1:])
        else:
            return max(self.longestCommonSubsequence(text1, text2[1:]), self.longestCommonSubsequence(text1[1:], text2))
```

但是这种方法最终显示超时了。我们也知道，使用动态规划时，可能会出现重复解的问题，在一些比较极端的情况下会带来很大的时间开销。比如以下这种情况：

很显然，我们可以看到bcd,efg的最长公共子串这个问题被重复计算了，当然，这也是动态规划的另外一个很重要的性质，重叠子问题，我们可以利用一个二维数组来存储这些中间结果，二维数组如下：


代码如下：

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        table = [[0 for i in range(len(text1))] for j in range(len(text2))]
        if text1[0] == text2[0]:
            table[0][0] = 1
        for i in range(1, len(text1)):
            if text2[0] == text1[i]:
                table[0][i] = 1
            else:
                table[0][i] = table[0][i-1]
        
        for j in range(1, len(text2)):
            if text1[0] == text2[j]:
                table[j][0] = 1
            else:
                table[j][0] = table[j-1][0]
        
        for i in range(1,len(text2)):
            for j in range(1,len(text1)):
                if text2[i] == text1[j]:
                    table[i][j] = 1 + table[i-1][j-1]
                else:
                    table[i][j] = max(table[i-1][j], table[i][j-1])
                    
        return table[-1][-1]
```
