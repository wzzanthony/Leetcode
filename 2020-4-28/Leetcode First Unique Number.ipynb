# Leetcode First Unique Number

这应该是leetcode中的一道新题，在leetcode官方题库中还未出现

题目描述：




解题思路：
这一题并不涉及任何复杂的算法，只需要按照接口的要求实现对应的功能即可，但是，我们应该尽可能的降低时间复杂度。其实我们并不需要把整个数组存储下来。通过观察，我们可以发现，在数组上的操作只有添加元素，而没有删除元素。所以我们一开始只需要存储一个独立数组（该数组中仅仅包含独立的元素，但是需要保证按照顺序存放），add操作时，我们判断添加的元素是否在独立数组中，在的话，删除该元素，不在的话，则将其添加到数组末尾。show操作的话，直接返回数组的第一个元素即可。

一开始，通过字典来生成独立数组，字典的key为对应的元素，value为其在原数组中的出现次数。之后，筛选出value为1的元素存入到独立数组中即可。add操作只需判断新添加的元素是否在dict中，不在的话，则将其添加到独立数组和字典中。


代码如下：

```python
class FirstUnique:

    def __init__(self, nums: List[int]):
        self.unique_arr = []
        self.dic = dict()
        for item in nums:
            if item in self.dic:
                self.dic[item] += 1
            else:
                self.dic[item] = 1
        for item in nums:
            if self.dic[item] == 1:
                self.unique_arr.append(item)

    def showFirstUnique(self) -> int:
        if len(self.unique_arr) == 0:
            return -1
        return self.unique_arr[0]

    def add(self, value: int) -> None:
        if value in self.dic:
            if self.dic[value] == 1:
                self.unique_arr.remove(value)
            self.dic[value] += 1
        else:
            self.dic[value] = 1
            self.unique_arr.append(value)
```
