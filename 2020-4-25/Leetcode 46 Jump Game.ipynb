# Leetcode 55 Jump Game

题目链接：https://leetcode.com/problems/jump-game/


解题思路1：
首先这一题，一开始想到的解决方案就是动态规划，同时为了节约空间，想到了用数组保存计算出的结果，这样的时间复杂度应该为$$O(n^2)$$，空间复杂度为$$O(n)$$


使用一个与给定数组等长的数组来存储各个位置的信息（True 或者 False），True代表从当前位置可以到达数组末尾， False则代表不能。初始时将全部元素设置为False，最后一位设置为True。

更新原则：

从倒数第二位开始遍历，如果当前数值（即步幅）为0， 则将该位置更新为False。如果当前数值加上当前坐标超出或者等于最后一位的坐标，则可以将该位置更新为True。否则，将其更新为其步幅范围内所能达到结果相或的结果（大概可以理解为，只要他能走到一个可以走到终点的位置即可）

代码如下：

```python
class Solution:     
    def canJump(self, nums: List[int]) -> bool:
        ret = [False for i in range(len(nums))]
        ret[len(nums) - 1] = True
        for i in range(len(nums) - 2, -1, -1):
            if nums[i] == 0:
                ret[i] == False
            else:
                if nums[i] + i >= len(nums) - 1:
                    ret[i] = True
                else:
                    ret[i] = False
                    for j in range(1,nums[i]+1):
                        if ret[i+j] == True:
                            ret[i] = True
                            break
                    
        return ret[0]
```

解题思路2：

仔细思考解题思路1，我们可以想到，我们想知道能不能走到终点，只要能保证每次走到一个可以走到终点的中间结果就可以了。所以，我们也不需要用
数组来存储中间结果，我们只需要不停的记录这个中间结果，并且不停地将其更新即可。这样时间复杂度为$$O(n)$$，空间复杂度是常数。

代码如下：

class Solution:     
    def canJump(self, nums: List[int]) -> bool:
        left_point = len(nums) - 1
        for i in range(len(nums) - 2, -1, -1):
            if i + nums[i] >= left_point:
                left_point = i
                
        return left_point == 0
